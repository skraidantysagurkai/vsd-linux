```python
import pandas as pd
import json 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Nuskaitom duomenis

```python
with open('data_vectors.json', 'r') as f:
        records = [json.loads(line) for line in f]
        

# Labels for the 'content' field
content_labels = [
    '5_min_1',
    '5_min_2',
    '5_min_3',
    '5_min_4',
    '5_min_5',
    '5_min_6',
    '5_min_7',
    '5_min_8',
    '5_min_9',
    '5_min_10',
    'thirty_sec_avg_arg_count',
    'thirty_sec_success_rate',
    '30_sec_1',
    '30_sec_2',
    '30_sec_3',
    '30_sec_4',
    '30_sec_5',
    '30_sec_6',
    '30_sec_7',
    '30_sec_8',
    '30_sec_9',
    '30_sec_10',
    'pid',
    'e_1',
    'e_2',
    'e_3',
    'e_4',
    'e_5',
    'e_6',
    'e_7',
    'e_8',
    'e_9',
    'e_10'
]

# Build DataFrame
rows = []
for record in records:
    row = {'id': record['id'], 'target': record['target']}
    row.update(dict(zip(content_labels, record['content'])))
    rows.append(row)

df = pd.DataFrame(rows)
```

### Subalansuojam klases

```python
df_0 = df[df['target'] == 0].sample(n=55, random_state=42)
df_1 = df[df['target'] == 1]
df = pd.concat([df_0, df_1], ignore_index=True)

```

### boxplotai

```python
columns = ['thirty_sec_avg_arg_count', 'thirty_sec_success_rate']
titles = ['Vidutiniško argumentų skaičiaus 30 sek. lange pasiskirstymas pagal klasę', 
          'Vidutiniško sekmės rodiklio 30 sek. lange pasiskirstymas pagal klasę']
y_labels = ['Vidtuinis argumentų skaičius 30 sek. lange', 
            'Vidtuinis sekmės rodiklis 30 sek. lange']
palette = "Dark2"

# Create two separate plots
for i, column in enumerate(columns):
    # Create a new figure for each plot
    plt.figure(figsize=(8, 5))
    
    # Create the boxplot
    sns.boxplot(x='target', y=column, data=df, palette=palette)
    
    # Set titles and labels
    plt.title(titles[i])
    plt.xlabel('Klasė')
    plt.ylabel(y_labels[i])
    
    # Optional: Adjust layout
    plt.tight_layout()
    
    # Show the plot (or you can save it)
    plt.show()
```

### five min pairplot

```python
g = sns.pairplot(df[[
    '5_min_1',
    '5_min_2',
    '5_min_3',
    '5_min_4',
    '5_min_5',
    '5_min_6',
    '5_min_10', 
    'target']], 
    hue='target')

# Update the legend title
g._legend.set_title('Klasė')
plt.show()
```
